# This notebook illustrates the Gaussian Process Regression

```python
import numpy as np

from matplotlib import pyplot as plt
from matplotlib import rc

from InfoGrad import InfoGrad
from IPython.display import HTML

import pickle as pkl

%load_ext autoreload
%autoreload 2
```

```python
# The underlying function
def f(x):
    return -1+0.01*(x-4.9)*(x+5)*(x+0.5)*(x-3.5)
```

```python
def plot1D(mode,mu_hist,std_hist,X,noisy_Y,n_0,n_robot):
    full_X = np.linspace(-5,5,100)
    plt.rc('font',size=16)
    fig = plt.figure(dpi=200)
    
        
    mu = mu_hist[-1]
    std = std_hist[-1]

    plt.title(mode)
    
    plt.plot(full_X,f(full_X),label=r'$f(x)$',lw=3,color='b')

    plt.scatter(X,noisy_Y,marker='+',label=r'Samples',s=200,color='b')


    
    plt.plot(full_X,mu,label=r'$\hat{f}(x)$',lw=3,ls='dotted',color='b')
    plt.fill_between(full_X,mu+std,mu-std,alpha=0.3)

    # plt.axhline(0,lw = 3,ls="-.",label='Sea level')
    plt.axhline(0,lw = 1,ls="-.")
    plt.ylim(-4,1)
    plt.legend(loc='upper left')
    plt.tight_layout()
    
    plt.savefig('{}1D.png'.format(mode))
```

```python
mode = 'greedy'
with open('{}.pkl'.format(mode),'rb') as file:
    data = pkl.load(file)

plot1D(mode,**data)
```

```python
mode = 'InfoGrad'
with open('{}.pkl'.format(mode),'rb') as file:
    data = pkl.load(file)

plot1D(mode,**data)
```

```python
mode = 'randomWalk'
with open('{}.pkl'.format(mode),'rb') as file:
    data = pkl.load(file)

plot1D(mode,**data)
```
