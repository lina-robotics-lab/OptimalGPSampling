```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = -(np.exp(-2*(X+1.5)**2 - 0.3*(Y+1.5)**2-0.3*X*Y+0.4)+np.exp(-(X-0.5)**2 - (Y-1.2)**2-0.6*X*Y)+np.exp(-0.7*(X-1.5)**2 - 0.3*(Y+1.5)**2-0.*X*Y))
# Z=Z+(np.random.randn(len(Z.ravel()))*0.5).reshape(Z.shape)
fig,ax=plt.subplots(1,1,dpi=200)
cp = ax.contourf(X, Y, Z,levels=9)
fig.colorbar(cp) # Add a colorbar to a plot
# ax.set_title('Depth of an imaginary sea floor')
#ax.set_xlabel('x (cm)')
# ax.set_ylabel('y (cm)')
fig.savefig('imaginarySeafloor.png')
plt.show()
```

```python
fig,ax=plt.subplots(1,1,dpi=200)
cp = ax.contour(X, Y, Z,levels=2,colors='k',alpha=1)
fig.colorbar(cp) # Add a colorbar to a plot
ax.clabel(cp, fontsize=9, inline=True)
# ax.set_title('Depth of an imaginary sea floor')
#ax.set_xlabel('x (cm)')
# ax.set_ylabel('y (cm)')
fig.savefig('predictedSeafloor.png')
plt.show()
```

```python
fig,ax=plt.subplots(1,1,dpi=200)
ax.set_ylim((-3,3))
ax.set_xlim((-3,3))
cp = ax.contour(X, Y, Z,levels=3,colors='k',alpha=0.0)
fig.colorbar(cp)
# ax.set_title('Depth of an imaginary sea floor')
#ax.set_xlabel('x (cm)')
# ax.set_ylabel('y (cm)')
fig.savefig('blank.png')
plt.show()
```
